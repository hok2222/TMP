```python
#Template method
class Process:
    def start(self):
        self.prepare()
        self.execute()
    def prepare(self):
        print("Preparing the process...")
    def execute(self):
        pass
class Wash(Process):
    def execute(self):
        print("Washing the floor")
class Build(Process):
    def execute(self):
        print("Building the furniture")
washing_process = Wash()
washing_process.start()
building_process = Build()
building_process.start()
```
